```python
import pandas

df = pandas.read_csv("./data_dcinside/data_hand_labelled.csv", index_col=0)
```

```python
df
```

```python
f = df["disaster"] == True
```

```python
df.loc[f, "disasterKind"] = df.loc[f]["keyword"]
```

```python
df["disasterKind"] = df["disasterKind"].replace({
    "지진": "earthquake",
    "화재": "fire",
    "불남": "fire",
    "싱크홀": "sinkhole",
}, )
```

```python
df.loc[df["disasterKind"].isna(), "disasterKind"] = "none"
```

```python
df
```

```python
df
```

```python
df["merged"] = df["title"].map(lambda x: str(x)) + "\n" + df["content"].map(lambda x: str(x))
```

```python
transformed = df[["merged", "disasterKind"]]
```

```python
transformed
```

```python
transformed.to_csv("./data_dcinside/transformed.csv")
```
